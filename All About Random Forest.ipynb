# Random Forest

Importing all the necessery modules and packages

```python
import numpy as np
import pandas as pd

from mlxtend.plotting import plot_decision_regions
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_gaussian_quantiles
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

Creating a dummy dataset for exploring the Random Forest Algorithm

```python
X1, y1 = make_gaussian_quantiles(
    cov=2.0, n_samples=200, n_features=2, n_classes=2
)
X2, y2 = make_gaussian_quantiles(
    mean=(3, 3), cov=1.5, n_samples=200, n_features=2, n_classes=2
)
X = np.concatenate((X1, X2))
y = np.concatenate((y1, -y2 + 1))
sb.scatterplot(x=X[:,0], y=X[:,1], hue=y)
```

Splitting the dataset into training and testing sets

```python
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.25)
```

Training a decision tree 

```python
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
plot_decision_regions(X,y,clf=model)
```

```python
print("Accuracy of Decision Tree ::",accuracy_score(y_test, y_pred))
```

Training a Random Forest

```python
model = RandomForestClassifier(n_estimators=300) #n_estimators is taking number of tree train in the random forest
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
plot_decision_regions(X,y,clf=model)
```

```python
print("Accuracy of Random Forest ::",accuracy_score(y_test, y_pred))
```
