```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from sklearn.datasets import make_circles, make_moons
import plotly.express as py
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Creating out own dataset
X, y = make_moons(n_samples=500, noise=0.3)
py.scatter(x=X[:,0],y=X[:,1], color=y)
```

```python
# Splitting the data into train test sets

x_train, x_test, y_train, y_test = train_test_split(X,y, train_size=0.8)
```

```python
x_train.shape
```

# Early stopping Example:

```python
# Constructing a deep neural sequencial model

model = Sequential()

model.add(Dense(250, activation='relu', input_dim=2))
model.add(Dense(100, activation='relu'))
model.add(Dense(1, activation=keras.activations.sigmoid))
```

```python
# Compiling the model
model.compile(optimizer=keras.optimizers.Adam(), loss=keras.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
# Training the model for 2 thousand epoch and 
history = model.fit(x_train,y_train,epochs=500, validation_data=(x_test,y_test))
```

```python
data = history.history

loss = data['loss']
val_loss = data['val_loss']

plt.plot(loss)
plt.plot(val_loss)
plt.legend(['loss', 'val_loss'])
```

#### Now applying Early Stopping

```python
# Early stopping 
callback = EarlyStopping(
    monitor="val_loss",
    min_delta=0,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
)
```

```python
# Constructing the same model and train it with EarlyStopping

model = Sequential()

model.add(Dense(250, input_dim=2, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(1, activation=keras.activations.sigmoid))
```

```python
# Compiling the model
model.compile(optimizer=keras.optimizers.Adam(), 
              loss=keras.losses.BinaryCrossentropy(), 
              metrics=['accuracy'])
```

```python
# Training the model with ErlyStopping
history = model.fit(x_train, y_train, epochs=500,  validation_data=(x_test,y_test),callbacks=callback)
```

```python
data = history.history

loss = data['loss']
val_loss = data['val_loss']

plt.plot(loss)
plt.plot(val_loss)
plt.legend(['loss', 'val_loss'])
```

# Dropout Technique

```python
# Constructing a deep neural sequencial model with out dropout layer

model = Sequential()

model.add(Dense(250, activation='relu', input_dim=2))
model.add(Dense(250, activation='relu'))
model.add(Dense(250, activation='relu'))
model.add(Dense(250, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(1, activation=keras.activations.sigmoid))
```

```python
# Compiling the model
model.compile(optimizer=keras.optimizers.Adam(), loss=keras.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
# Training the model for 2 thousand epoch and 
history = model.fit(x_train,y_train,epochs=400, validation_data=(x_test,y_test))
```

```python
plt.figure(figsize=(10,8))
plot_decision_regions(X=x_test,y=y_test.ravel(),clf=model)
```

#### Using Dropout 

```python
# Constructing a deep neural sequencial model with out dropout layer

model = Sequential()

model.add(Dense(250, activation='relu', input_dim=2))
model.add(Dropout(0.7))
model.add(Dense(250, activation='relu'))
model.add(Dropout(0.8))
model.add(Dense(250, activation='relu'))
model.add(Dropout(0.6))
model.add(Dense(250, activation='relu'))
model.add(Dropout(0.8))
model.add(Dense(100, activation='relu'))
model.add(Dropout(0.7))
model.add(Dense(1, activation=keras.activations.sigmoid))
```

```python
# Compiling the model
model.compile(optimizer=keras.optimizers.Adam(), loss=keras.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
# Training the model for 2 thousand epoch and 
history = model.fit(x_train,y_train,epochs=400, validation_data=(x_test,y_test))
```

```python
plt.figure(figsize=(10,8))
plot_decision_regions(X=x_test,y=y_test.ravel(),clf=model)
```

```python
# Constructing a deep neural sequencial model with regularization

model = Sequential()

model.add(Dense(250, activation='relu', input_dim=2))
model.add(Dense(250, activation='relu', kernel_regularizer=keras.regularizers.L2()))
model.add(Dense(250, activation='relu', kernel_regularizer=keras.regularizers.L2()))
model.add(Dense(250, activation='relu', kernel_regularizer=keras.regularizers.L2()))
model.add(Dense(100, activation='relu', kernel_regularizer=keras.regularizers.L2()))
model.add(Dense(1, activation=keras.activations.sigmoid))
```

```python
# Compiling the model
model.compile(optimizer=keras.optimizers.Adam(), loss=keras.losses.BinaryCrossentropy(), metrics=['accuracy'])
```
